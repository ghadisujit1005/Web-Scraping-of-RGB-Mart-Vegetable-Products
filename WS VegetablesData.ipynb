```python
Url = 'https://rgbmart.com/'
print(type(Url))
```

```python
responce = requests.get(Url)
print(responce)
```

```python
Soup = BeautifulSoup(responce.text,'html.parser')
print(Soup)
```

```python
# extract title

Soup.title.text
```

```python
# return h1 heading text

Soup.h1.text
```

```python
# return all (h1) type heading

heading = Soup.find_all("h1")  
print(heading)
```

```python
Soup.h2.text
```

```python
heading2 = Soup.find_all("h2")
print(heading2)
```

```python
# list compression
H =[i.text for i in heading]
print(H)
```

```python
H2 = [i.text for i in heading2]
print(H2)
```

```python
# extract product_id

import re

product_ids = []

ids = Soup.find_all("span", {"class":"screen-reader-text"})

for s in ids:
    span_id = s.get("id", "")
    match = re.search(r"describedby_(\d+)", span_id)
    if match:
        product_ids.append(match.group(1))

# Remove duplicates
product_ids = list(set(product_ids))

# Print all product IDs
for product_id in sorted(product_ids):
    print(product_id)
```

```python
# Extract Product_Name

products =Soup.find_all("h2", {"class": "woocommerce-loop-product__title"})

products1 =[i.text for i in products]
print(products1)
```

```python
# Extract Price

price = Soup.find_all("span",{"class": "woocommerce-Price-amount amount"})

price1=[ i.text for i in price]
print(price1)
```

```python
# Extract quantity

quantity = Soup.find_all("span",{"class":"qty_type"})

quantity1 =[i.text for i in quantity]
print(quantity1)
```

```python
# Extract all Product link

product_links = []

# Find all product <a> tags
links = Soup.find_all("a", {"class":"woocommerce-LoopProduct-link"})

for a in links:
    href = a.get("href")
    if href and href.startswith("http"):
        product_links.append(href)

# Remove duplicates
product_links = list(set(product_links))

# Print links
for link in sorted(product_links):
    print(link)
```

```python
# Create dictionary

dict ={"Product ID" : product_id ,"Product Name" : products1, "Quantity" : quantity1, "Price" : price1, "Product link" : link}
```

```python
# create dataframe using pandas

import pandas as pd
df = pd.DataFrame(dict)
df
```

```python
# to save data from dataframe df to permanent file. csv in default path
# use inbuilt method to_csv()

df.to_csv('VegetablesData.csv', index=False)
```

# import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Scatter plot:
# x axis (quantity) is independant variable and y axis(price) is dependant variable
# scatterplot() : inbuilt function of matplotlib.pyplot

plt.scatter(x = 'Quantity',y = 'Price',data = df,c='blue',marker ='*') 
plt.title('Quantity vs Price') #create title
plt.xlabel('Quantity')
plt.ylabel('Price (₹)')
plt.show
```

```python
# Distribution of Quantity (counts Quantity)

df['Quantity'].value_counts().plot(kind = "bar")
count = df['Quantity'].value_counts()
plt.yticks(count) # present actual values
plt.show()
```

```python
# barplot - Product name vs prce



plt.figure(figsize=(10,6))
sns.countplot(y=df["Price"], data = df, order = df["Price"].value_counts().iloc[1:6].index)
plt.title("Count Prices")
plt.show()
```

```python
# CountPlot - Product per quantity type

plt.figure(figsize=(8,4))
sns.countplot(x="Quantity",data=df, )
plt.title("Quantity Type Distribution")
plt.xlabel("Quantity")
plt.ylabel("Count")
plt.show()
```

```python
# Histgram - Price Distribution

plt.figure(figsize=(8,5))
sns.histplot(df["Price"], kde=True) #Kernel Density Estimation
# It draws a smooth curved line that represents the probability distribution of the data.
plt.title("Price Distribution")
plt.xlabel("Price (₹)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot - price spread

sns.boxplot(y = 'Price',data = df)
plt.title("Price Spread of Product")
plt.ylabel("Price (₹)")
plt.show()
```
